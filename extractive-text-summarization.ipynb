```python
def text_summarizer(text):
    doc=nlp(text)
    tokens =[token.text for token in doc]
    
    word_frequencies={}
    for word in doc:
        if word.text not in STOP_WORDS:
            if word.text not in punctuation:
                if word.text not in word_frequencies:
                    word_frequencies[word.text]=1
                else:
                    word_frequencies[word.text]=word_frequencies[word.text]+1
                    
    
    max_frequencies=max(word_frequencies.values())
    
    for word in word_frequencies.keys():
        word_frequencies[word]= word_frequencies[word]/max_frequencies
    
    sents=[sent for sent in doc.sents]
    
    sent_scores={}
    for sent in sents:
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sent_scores.keys():
                    sent_scores[sent]=word_frequencies[word.text.lower()]
                else:
                    sent_scores[sent]+=word_frequencies[word.text.lower()]
                    
    
    x=int(len(sents)*0.2)
    summary=nlargest(x,sent_scores,key=sent_scores.get)
    final_summary=[word.text for word in summary]
    Summary=' '.join(final_summary)
    
    return Summary    
```

## Now performing the Extractive text summarization using TFIDF method

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def text_summarizer_tfidf(text,text__doc):
    v=TfidfVectorizer()
    v.fit(text)

    all_feature_names = v.get_feature_names_out()

    word_frequencies={}
    for word in all_feature_names:
        if word not in STOP_WORDS:
            if word not in punctuation:
                if word not in word_frequencies:
                    indx = v.vocabulary_.get(word)
                    idf_score = v.idf_[indx]
                    word_frequencies[word]=idf_score
    
    
    doc=nlp(text__doc)
    sents=[sent for sent in doc.sents]
    
    sent_scores={}
    for sent in sents:
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sent_scores.keys():
                    sent_scores[sent]=word_frequencies[word.text.lower()]
                else:
                    sent_scores[sent]+=word_frequencies[word.text.lower()]
                    
    
    x=int(len(sents)*0.2)
    summary=nlargest(x,sent_scores,key=sent_scores.get)
    final_summary=[word.text for word in summary]
    Summary=' '.join(final_summary)
    
    return Summary    
```

## Extractive text summarization using sentence similarity methods

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_similarity_scores(sentences):
    # Tokenize sentences
    tokenized_sentences = [nltk.word_tokenize(sentence) for sentence in sentences]

    # Convert sentences to lowercase
    tokenized_sentences = [[token.lower() for token in sentence] for sentence in tokenized_sentences]

    # Flatten tokenized sentences
    flattened_sentences = [' '.join(tokens) for tokens in tokenized_sentences]

    # Create TF-IDF vectorizer
    vectorizer = TfidfVectorizer()

    # Compute TF-IDF matrix
    tfidf_matrix = vectorizer.fit_transform(flattened_sentences)

    # Calculate cosine similarity matrix
    similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)

    # Assign similarity scores to each sentence
    similarity_scores = []
    for i in range(len(sentences)):
        similarity_scores.append(sum(similarity_matrix[i]) / len(sentences))

    return similarity_scores


# Example usage
paragraph = "Varun is a nice person.Varun helps others. Varun is a kind and generous person. This world is quite beautiful. We should care about our nature."
sentences = nltk.sent_tokenize(text_doc)

similarity_scores = calculate_similarity_scores(sentences)

# Print the sentences and their corresponding similarity scores
for i in range(len(sentences)):
    print("Sentence:", sentences[i])
    print("Similarity Score:", similarity_scores[i])
    print()
```

```python
def text_summarize_similarity_scores(text):
    
    sentences = nltk.sent_tokenize(text)

    similarity_scores = calculate_similarity_scores(sentences)
    
    max_score=max(similarity_scores)
    min_score=min(similarity_scores)
    
    avg_score=(min_score+max_score)/2;
    
    text_summary=[]
    for score in range(len(similarity_scores)):
        if similarity_scores[score]>(avg_score+0.02) or similarity_scores[score]<(avg_score-0.02):
            text_summary.append(sentences[score])
            
    final_summary=" ".join(text_summary)
    
    return final_summary
```

## Extractive text summarization using Clustering methods

#### Using the cluster based algorithms for grouping similar clusters and then chosing or selecting the sentences from each clusters for the required summary 

```python
from sklearn.cluster import KMeans 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def kmeans_text_summarization(text, num_clusters):
    # Tokenize the text into sentences
    sentences = nltk.sent_tokenize(text)

    # TF-IDF vectorization
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(sentences)

    # Perform K-means clustering
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(tfidf_matrix)

    # Get cluster assignments for each sentence
    cluster_labels = kmeans.labels_

    # Select representative sentences from each cluster
    summary_sentences = []
    for cluster_id in range(num_clusters):
        cluster_sentences = [sentences[i] for i, label in enumerate(cluster_labels) if label == cluster_id]
        representative_sentence = max(cluster_sentences, key=lambda sentence: vectorizer.transform([sentence]).sum())
        summary_sentences.append(representative_sentence)

    # Construct the summary
    summary = ' '.join(summary_sentences)

    return summary
```
